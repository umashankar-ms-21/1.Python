# odd or even number finding and excute with definition variable name:

```python
def oddEven(): # If haven't mentioned any parameter within the paranthesis then it is calles as "NON-PARAMETERISED OR NON-ARGUMENTS"
    num=int(input("Enter the number:"))
    if((num%2)==1):
        print("Odd Number")
        message="Odd Number"
    else:
        print("Even Number")
        message="Even Number"
    return message
```

```python
oddEven()
```

```python
# the ouput has been assigned in the "message" variable:
message=oddEven()
```

```python
# message variable excuted with another if-else & print statement:
if(message=="Even Number"):
    print("The given number is even number")
else:
    print("The given number is odd number")
```

```python
# if the definition doesn't have the return variable call then what would be the output:

def oddEvenNoR():
    num=int(input("Enter the number:"))
    if((num%2)==1):
        print("Odd Number")
        message="Odd Number"
    else:
        print("Even Number")
        message="Even Number"
```

```python
oddEvenNoR()
```

```python
messageNoR=oddEvenNoR()
```

```python
print(messageNoR)
```

```python
# Return call not assigned in the definition however excuted with another if-else & print statement:
if(messageNoR=="Even number"):
    print("The given number is even number")
else:
    print("Return call not assigned in the definition hence the output is NONE")
```

```python
def BMI():
    BMI=float(input("Enter the BMI Index:"))
    if(BMI<18.5):
        print("Underweight")
        message="Underweight"        
    elif(BMI<24.9):
        print("Normal")
        message="Normal"
    elif(BMI<29.9):
        print("Overweight")
        message="Overweight"
    elif(BMI>=30):
        print("Very Overweight")
        message="Very Overweight"
    return message
```

```python
bmi=BMI()
```

# If have mentioned any parameter within the paranthesis then it is called as "PARAMETERISED OR ARGUMENTS" 

```python
def addtion(num1,num2):
    add=num1+num2
    return add
```

```python
addtion(num1,num2) # If the statement excuted before defined variable then wouldnn't come ouput instead of "NameError"
```

```python
addtion(4,5) # instead of call with assigned variables then would be an output
```

```python
def sub(num1,num2):
    sub=num1-num2
    return sub
```

```python
sub(2,3,4) # positional arguments must be two(num1,num2), if exceeds then would be "TypeError" 
```

```python
sub(2,3) # positional arguments accepted for two(num1,num2)
```
