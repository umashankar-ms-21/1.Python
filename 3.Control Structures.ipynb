```python
age=int(input("Enter your age :"))
```

```python
# if-elif meaning : single input "if" condition is multiple then "if-elif" will apply:

if(age<18):
    print("Children")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior Citizen")
```

```python
-----Important Note-------
# single input & single condition = "if"
# single input & two conditions = "if-else"
# single input & multiple conditions = "if-elif"
# however, the single result printed for continuos input, what is the reason for that?
# since the memory saved only the last input and also excutes the print result for last input.
```

```python
lists=[25,45,55,79,18,12,35,49,65]
```

```python
for age in lists:
    if(age<18):
        print("Children")
    elif(age<35):
        print("Adult")
    elif(age<59):
        print("Citizen")
    else:
        print("Senior Citizen")
```

```python
------Important Note-------
# when the multiple defined variables stored in the data then apply on the set of if-elif conditional procedure.
# "if-elif" is used to get the set of conditional procedure print results at the same time on "for loop" inputs.
# here single input couldn't provide result on "if-elif" condition.
```
