```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
age=int(input("age:"))

if(age<18):
    print("Children")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior Citizen")
```
