```python
# Create a class and function, and list out the items in the list
```

```python
AI=["Subfileds","Number","Brain","Eye","Hardware","Speech","Text"]

class SubfieldsInAI():
    def Subfields():
        for Subfields in AI:
            if(Subfields=="Subfileds"):
                print("Sub-fields in AI are:")                     # without assignment vaiable            
            elif(Subfields=="Number"):
                print("Machine Learning")
            elif(Subfields=="Brain"):
                print("Neural Networks")
            elif(Subfields=="Eye"):
                print("Vision")
            elif(Subfields=="Hardware"):
                print("Robotics")
            elif(Subfields=="Speech"):
                print("Speech Processing")
            elif(Subfields=="Text"):
                print("Natural Language Processing")
            else:
                print()
```

```python
SubfieldsInAI.Subfields()
```

```python
# Create a function that checks whether the given number is Odd or Even
```

```python
class OddEven():
    def OddEven():
        num=int(input("Enter a number :"))
        if((num%2)==0):
            print(num,"is Even number")
        else:
            print(num,"is Odd number")
```

```python
OddEven.OddEven()
```

```python
# Create a function that tells elegibility of marriage for male and female according to their age limit like 21 for male and 18 for female
```

```python
class ElegiblityForMarriage():
    def Elegible():
        gender=input("Your Gender :")
        age=int(input("Your Age :"))
        if(gender=="Male") & (age>=21):         # here used "&" operator instead of word "and"
            print("ELIGIBLE")
        elif(gender=="Female") and (age>=18):   # here used word "and" instead of "&" operator, both can be used.
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")
```

```python
ElegiblityForMarriage.Elegible()
```

```python
# calculate the percentage of your 10th mark#
```

```python
class FindPercent():
    def percentage():
        sub1=int(input("Subject1= "))
        sub2=int(input("Subject2= "))
        sub3=int(input("Subject3= "))
        sub4=int(input("Subject4= "))
        sub5=int(input("Subject5= "))
        Total=(sub1+sub2+sub3+sub4+sub5)
        print("Total :", Total)
        Percentage=Total/5
        print(f"Percentage :{Percentage:.15f}")
```

```python
FindPercent.percentage()
```

```python
# print area and perimeter of triangle using class and functions
```

```python
class triangle():
    def triangle():
        Height=int(input("Height :"))
        Breadth=int(input("Breadth :"))
        triangleArea=(Height*Breadth)/2
        print("Area formula: (Height*Breadth)/2")
        print("Area of Triangle :", triangleArea)
        Height1=int(input("Height1 :"))
        Height2=int(input("Height2 :"))
        Breadth1=int(input("Breadth1 :"))
        perimeterArea=Height1+Height2+Breadth1
        print("Perimeter formula: Height1+Height2+Breadth1")
        print("Perimeter of Triangle :", perimeterArea)
```
